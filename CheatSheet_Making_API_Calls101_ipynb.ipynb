### Read the CSV with demo URLs into a dictionary

```
import csv
import requests
```

```
data_file = csv.DictReader(open(file_names[0]))
```

Reading into a dictionary eliminates duplicates

```
themes = dict()

for row in data_file:

  theme = row["theme"]

  link = row["demo-url"]

  themes[link] = theme
```

```
len(themes)
```

### Making the API call

```
from urllib.parse import quote

def get_psi(url, category, strategy, psi_key):

  encoded_url = quote(url)
  print(encoded_url)

  api_url = f'https://pagespeedonline.googleapis.com/pagespeedonline/v5/runPagespeed?category={category}&strategy={strategy}&url={encoded_url}&key={psi_key}'

  r = requests.get(api_url, headers={'Accept': 'application/json'})

  if r.status_code == 200:

    return r.json()

  else:
    print(api_url)
    print(r.status_code)
    print(url+" failed")
```

### Testing the API call works as expected

Prepare the parameters

```
url = "https://story-theme.myshopify.com/"
strategy = "MOBILE"
category = "SEO"
```

Capture the JSON response

```
response = get_psi(url, category, strategy, psi_key)
```

### Save the JSON responses to disk

```
import json

for link, theme in themes.items():
  #OrderedDict([('link', 'https://themes.shopify.com/themes/express/styles/bistro'), ('theme', 'Express')])
  filename = f"{theme}.json"
  
  strategy = "MOBILE"
  category = "SEO"

  response = response = get_psi(link, category, strategy, psi_key)

  with open(filename, "w") as f:
    print(f"saved {filename}")
    f.write(json.dumps(response))
```
