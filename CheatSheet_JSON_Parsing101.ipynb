### Getting metrics and recommendations from the parsed JSONs



```
from jsonpath_ng import jsonpath, parse
import json
from urllib.parse import urlparse
```

```
def get_value(json_path, data):
    jsonpath_tree = parse(json_path)

    value = [match.value for match in jsonpath_tree.find(data)]

    def get_first(x):

      if len(x) > 0:
        return x[0]

    return get_first(value)
```

```


def get_metrics(json_path, filename):

  #Read JSON file
  with open(filename) as f:
    data = json.load(f)

    return get_value(json_path, data)

```

```
get_metrics("$.lighthouseResult.categories.seo.score", "Debut.json")
```

```
audits = ["is-crawlable",
"meta-description",
"hreflang",
"canonical",
"document-title",
"image-alt",
"structured-data",
"tap-targets"]
```

```
from collections import defaultdict

theme_audits = defaultdict(list)
```

```
for theme_file in files:

  theme_name = theme_file.replace(".json", "")

  seo_score = get_metrics("$.lighthouseResult.categories.seo.score", theme_file)

  theme_audits[theme_name].append({"seo_score": seo_score})

  for audit in audits:
    #print(audit)
    audit_title = f"$.lighthouseResult.audits.{audit}.title"
    audit_description = f"$.lighthouseResult.audits.{audit}.description"
    audit_score = f"$.lighthouseResult.audits.{audit}.score"

    theme = dict()

    theme["audit"] = audit

    theme["audit_title"] = get_metrics(audit_title, theme_file)

    theme["audit_description"] = get_metrics(audit_description, theme_file)

    theme["audit_score"] = get_metrics(audit_score, theme_file)

    theme_audits[theme_name].append(theme)
```
